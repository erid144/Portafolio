```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.regularizers import l2
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Cargar el dataset
ruta = "/content/flight_df - flight_df.csv"  # Ruta al dataset
df = pd.read_csv(ruta)
df.columns = df.columns.str.replace('-', '_', regex=True).str.replace('/', '_', regex=True).str.replace(' ', '_', regex=True).str.upper()

# Seleccionar las características relevantes para el análisis
features = ['AGE', 'CLASS', 'FLIGHT_DISTANCE', 'INFLIGHT_WIFI_SERVICE', 'DEPARTURE_ARRIVAL_TIME_CONVENIENT',
            'EASE_OF_ONLINE_BOOKING', 'GATE_LOCATION', 'FOOD_AND_DRINK', 'ONLINE_BOARDING', 'SEAT_COMFORT',
            'INFLIGHT_ENTERTAINMENT', 'ON_BOARD_SERVICE', 'LEG_ROOM_SERVICE', 'BAGGAGE_HANDLING', 'CHECKIN_SERVICE',
            'INFLIGHT_SERVICE', 'CLEANLINESS', 'DEPARTURE_DELAY_IN_MINUTES', 'ARRIVAL_DELAY_IN_MINUTES',
            'GENDER_MALE', 'LOYAL_CUSTOMER', 'BUSINESS_TRAVEL']

# Preprocesamiento de los datos
X = df[features]
y = df['SATISFIED']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Aplicar ACP para reducir el número de características
pca = PCA(n_components=0.90)
X_pca = pca.fit_transform(X_scaled)
n_columns_after_pca = X_pca.shape[1]
variance_retained = np.sum(pca.explained_variance_ratio_)
print("El número de componentes es", n_columns_after_pca, "de", X_scaled.shape[1])

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)

# Crear modelos secuenciales
model1 = Sequential([
    Dense(32, activation='relu', input_shape=(n_columns_after_pca,)),
    Dense(1, activation='sigmoid')
])

model2 = Sequential([
    Dense(64, activation='relu', input_shape=(n_columns_after_pca,)),
    Dropout(0.4),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

model3 = Sequential([
    Dense(128, activation='relu', input_shape=(n_columns_after_pca,)),
    Dropout(0.6),
    Dense(64, activation='relu'),
    Dropout(0.4),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compilar los modelos
model1.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model2.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model3.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])




```
